```python
import pandas as pd
import numpy as np
```

```python
df_teams = pd.read_csv('baseballdatabank-2022.2/core/Teams.csv')
```

```python
df_teams['1B'] = df_teams['H'] - df_teams['2B'] - df_teams['3B'] - df_teams['HR']
```

```python
df_teams.insert(17, '1B', df_teams.pop('1B'))
```

```python
#Team runs created

df_teams['PrdTeamRunsCreated'] = (df_teams['H'] + df_teams['BB'] + df_teams['HBP']) * (df_teams['1B'] + (2*df_teams['2B']) + (3*df_teams['3B']) + (4*df_teams['HR'])) / (df_teams['AB'] + df_teams['BB'] + df_teams['HBP'])
```
