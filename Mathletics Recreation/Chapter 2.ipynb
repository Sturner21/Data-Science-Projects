# Personal Thoughts

What are the top 10 seasons with the highest RCPG of all season and who holds them?

```python
df_batting.sort_values('RCPG', ascending=False)
```

Obviously this is a flawed system since people who have played in less than 10 games have several hundred runs created per game. Let's filter out anyone with less than 300 AB's. It is fairly arbitrary why I am picking 300. The Mathletics authors use an example of two fictitious players to show the benefits of judging players by RCPG instead of raw predicted runs created because it favors players who have more AB's. In this example they gave the player with less AB's 300 which is why I picked 300.

In the future finding a better way to calculating runs created per game that doesn't favor players who play in very little games would be and would be an interesting problem.

```python
df10 = df_batting.query("AB >= 300").sort_values('RCPG', ascending=False)[:10].copy()
df10
```

```python
df10['playerID'].to_list()
df_people[df_people['playerID'].isin(df10['playerID'].tolist())][['playerID', 'nameFirst', 'nameLast']]
```

As we can see the top 3 season are held by Barry Bonds in 2004, 2002, 2001, and he held the 5th highest runs created per games in 2003. Ted Williams holds 2 of the top 5 at 4th and 7th. Mickey Mantle holds two lower spots at 6th and 10th. Frank Thomas and Jeff Bagwell have mentions too at 8th and 9th respectively.
